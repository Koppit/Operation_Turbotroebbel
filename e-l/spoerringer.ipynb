```python

def fetch_studyprograms() -> list[dict]:
    conn = get_connection()
    try:
        cursor = conn.cursor(dictionary=True)
        cursor.execute(
            "SELECT * FROM study_programs;" 
        )
        rows = cursor.fetchall()
        
        print("Rows fetched:", len(rows) )
        cursor.close()
        return rows
    finally:
        conn.close()

fetch_studyprograms()
```
